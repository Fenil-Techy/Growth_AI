```python
import os
import traceback
```

```python
from langchain_ollama import ChatOllama
from langchain.prompts import ChatPromptTemplate, HumanMessagePromptTemplate
```

```python
llm=ChatOllama(model="llama3.1:8b")
```

```python
TEMPLATE = """
You are an educational roadmap generator.

ðŸ“Œ Context: {context}  
ðŸ“Œ Target Time Period: {time_period}  
ðŸ“Œ Prior Knowledge: {learnings}  

Generate a personalized, beginner-friendly roadmap covering the specified context from scratch to advanced level, within the given time period.  
If the user already knows certain topics (mentioned in Prior Knowledge), skip or shorten those parts accordingly.

âœ… The roadmap should:
- Be divided into weekly (or appropriate) milestones.
- Include specific topics/concepts to cover at each step.
- Suggest free and reliable learning resources where possible.
- Be clear, actionable, and easy to follow for a beginner.

Respond in a structured, readable format (markdown preferred).
"""
```

```python
prompt=ChatPromptTemplate.from_messages([
    HumanMessagePromptTemplate.from_template(TEMPLATE)
])
```

```python
context="python"
time_period="3-months"
learnings="nothing"
```

```python
generator=prompt | llm
```

```python
chain=generator.invoke({"context":context,"time_period":time_period,"learnings":learnings})
```

```python
print(chain.content)
```
